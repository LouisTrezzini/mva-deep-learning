# A Regression Problem

```python
from keras.utils import np_utils


def reorder(Y):
    Y = Y.reshape(-1, 3, 2)
    
    Yp = Y.copy()
    
    for i in range(len(Y)):
        y = Y[i]
        # Sort by x-coordinate
        ordered = np.argsort(y[:, 0])

        Yp[i] = Yp[i, ordered, :]
    
    return Yp


X_train, Y_train = generate_dataset_regression(3000, 20)
Y_train = reorder(Y_train)
X_train = X_train.reshape(-1, 72, 72, 1)
```

```python
visualize_prediction(X_train[0], Y_train[0])
```

```python
X_test, Y_test = generate_test_set_regression()
Y_test = reorder(Y_test)
X_test = X_test.reshape(-1, 72, 72, 1)
```

## Answer

```python
from keras import optimizers

model = Sequential([
    Reshape((72, 72, 1), input_shape=(72, 72, 1)),
      
    Conv2D(32, (5, 5), padding='same'),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(16, (5, 5), padding='same'),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(8, (5, 5), padding='same'),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
           
    Flatten(),
    
    Dropout(0.3),
    Dense(6),
    Activation('relu'),
    Reshape((3, 2)),
])

model.summary()

model.compile(
    loss='mse',
    optimizer='adam',
)

model.fit(X_train, Y_train, batch_size=32, epochs=50, verbose=1, validation_split=0.15, shuffle=True)
```

# Image Denoising

```python
X_train, Y_train = generate_dataset_denoising(300)
X_train /= 255
Y_train /= 255
```

```python
print(X_train[0].shape, Y_train[0].shape, X_train.min(), X_train.max())

plt.imshow(X_train[0].reshape(72, 72), cmap='gray')
plt.show()
plt.imshow(Y_train[0].reshape(72, 72), cmap='gray')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Reshape, BatchNormalization, UpSampling2D

model = Sequential([
    Reshape((72, 72, 1), input_shape=(5184,)),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),

    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),
    
    Conv2D(1, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('sigmoid'),
    
    Reshape((5184,)),
])

model.summary()

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
)
model.fit(X_train, Y_train, batch_size=32, epochs=30, verbose=1)
```

```python
X_test, Y_test = generate_dataset_denoising(20)
X_test /= 255
Y_test /= 255
model.evaluate(X_test, Y_test)
Y_pred = model.predict(X_test)

for k in range(5):
    plt.subplot(1, 3, 1)
    plt.imshow(X_test[k].reshape(72, 72), cmap='gray')
    plt.title('Original image')
    plt.subplot(1, 3, 2)
    plt.imshow(Y_test[k].reshape(72, 72), cmap='gray')
    plt.title('Noisy image')
    plt.subplot(1, 3, 3)
    plt.imshow(Y_pred[k].reshape(72, 72), cmap='gray')
    plt.title('Denoised image')
    plt.show()
```

# Image Denoising

```python
X_train, Y_train = generate_dataset_denoising(300)
```

```python
print(X_train[0].shape, Y_train[0].shape)

plt.imshow(X_train[0].reshape(72, 72), cmap='gray')
plt.show()
plt.imshow(Y_train[0].reshape(72, 72), cmap='gray')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Reshape, BatchNormalization, UpSampling2D

model = Sequential([
    Reshape((72, 72, 1), input_shape=(5184,)),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2, 2), padding='same'),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),

    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),
    
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    UpSampling2D((2, 2)),
    
    Conv2D(1, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('sigmoid'),
    
    Reshape((5184,)),
])

model.summary()

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
)
model.fit(X_train, Y_train, batch_size=32, epochs=30, verbose=1)
```

```python
X_test, Y_test = generate_dataset_denoising(20)
model.evaluate(X_test, Y_test)
Y_pred = model.predict(X_test)

plt.imshow(X_test[0].reshape(72, 72), cmap='gray')
plt.show()
plt.imshow(Y_test[0].reshape(72, 72), cmap='gray')
plt.show()
plt.imshow(Y_pred[0].reshape(72, 72), cmap='gray')
plt.show()
```
